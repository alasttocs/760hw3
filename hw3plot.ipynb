```python
import matplotlib.pyplot as plt
import numpy as np
x = [0, 0, .25, .5, 1]
y = [0, 0.33333, 0.66667, 1,1]


plt.plot(x, y, marker='o')


plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC Curve')
plt.grid(True)

plt.savefig('5a.png')
plt.show()
```
